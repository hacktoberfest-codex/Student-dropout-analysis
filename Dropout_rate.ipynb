## **Model Building**

### Linear Regression

#### Training the Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

#### **Applying the model to make a prediction**

```python
y_model_train_pred = model.predict(x_train)
y_model_test_pred = model.predict(x_test)
```

```python
y_model_train_pred
```

```python
y_model_test_pred
```

#### Evaluate model performance

```python
y_train
```

```python
y_model_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

model_train_mse = mean_squared_error(y_train, y_model_train_pred)
model_train_r2 = r2_score(y_train, y_model_train_pred)

model_test_mse = mean_squared_error(y_test, y_model_test_pred)
model_test_r2 = r2_score(y_test, y_model_test_pred)
```

```python
model_results = pd.DataFrame(['Linear regression', model_train_mse, model_train_r2, model_test_mse, model_test_r2]).transpose()
model_results.columns = ['Method', 'Training MSE','Training R2','Test MSE','Test R2']
```

```python
model_results
```

### **Random Forest**

#### Training the model

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(max_depth=2, random_state=5)
model_rf.fit(x_train, y_train)
```

#### Applying the model to make a prediciton

```python
y_model_rf_train_pred = model_rf.predict(x_train)
y_model_rf_test_pred = model_rf.predict(x_test)
```

#### Evaluate model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

model_rf_train_mse = mean_squared_error(y_train, y_model_rf_train_pred)
model_rf_train_r2 = r2_score(y_train, y_model_rf_train_pred)

model_rf_test_mse = mean_squared_error(y_test, y_model_rf_test_pred)
model_rf_test_r2 = r2_score(y_test, y_model_rf_test_pred)
```

```python
model_rf_results = pd.DataFrame(['Linear regression', model_rf_train_mse, model_rf_train_r2, model_rf_test_mse, model_rf_test_r2]).transpose()
model_rf_results.columns = ['Method', 'Training MSE','Training R2','Test MSE','Test R2']
```

```python
model_rf_results
```

## **Model Comparison**

```python
data_models = pd.concat([model_results, model_rf_results], axis = 0)
data_models.reset_index(drop=True)
```
